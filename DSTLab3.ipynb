# Circular Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class CircularLinkedList:
    def __init__(self):
        self.head = None

    def insert_at_beginning(self, data):
        new_node = Node(data)
        if not self.head:
            new_node.next = new_node
            self.head = new_node
        else:
            n = self.head
            while n.next != self.head:
                n = n.next
            n.next = new_node
            new_node.next = self.head
            self.head = new_node

    def insert_after_element(self, prev_data, data):
        new_node = Node(data)
        if not self.head:
            print("List is empty. Cannot insert after element.")
            return
        n = self.head
        while n.next != self.head and n.data != prev_data:
            n = n.next
        if n.data != prev_data:
            print(f"Element {prev_data} not found. Cannot insert after element.")
            return
        new_node.next = n.next
        n.next = new_node

    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            new_node.next = new_node
            self.head = new_node
        else:
            n = self.head
            while n.next != self.head:
                n = n.next
            n.next = new_node
            new_node.next = self.head

    def delete_at_beginning(self):
        if not self.head:
            print("List is empty. Cannot delete from beginning.")
            return
        n = self.head
        while n.next != self.head:
            n = n.next
        if n == self.head:
            self.head = None
        else:
            n.next = self.head.next
            self.head = self.head.next

    def delete_by_value(self, value):
        if not self.head:
            print("List is empty. Cannot delete by value.")
            return
        n = self.head
        prev = None
        while n.next != self.head and n.data != value:
            prev = n
            n = n.next
        if n.data != value:
            print(f"Value {value} not found. Cannot delete by value.")
            return
        if prev:
            prev.next = n.next
        else:
            self.delete_at_beginning()

    def delete_at_end(self):
        if not self.head:
            print("List is empty. Cannot delete from end.")
            return
        n = self.head
        prev = None
        while n.next != self.head:
            prev = n
            n = n.next
        if prev:
            prev.next = self.head
        else:
            self.head = None

    def display(self):
        if not self.head:
            print("List is empty.")
            return
        n = self.head
        while True:
            print(n.data, end=" -> ")
            n = n.next
            if n == self.head:
                break
        print()


# Example usage:
circular_list = CircularLinkedList()

circular_list.insert_at_beginning(1)
circular_list.insert_at_beginning(0)
circular_list.insert_at_end(3)
circular_list.insert_after_element(1, 2)

print("Circular Linked List:")
circular_list.display()

circular_list.delete_at_beginning()
circular_list.delete_by_value(1)
circular_list.delete_at_end()

print("\nCircular Linked List after deletions:")
circular_list.display()
```
